```python
import pandas as pd


data = {'Country': ['Nigeria', 'France', 'China', 'USA', 'Brazil', 'Australia']}
df_clean = pd.DataFrame(data)


def result(country):
    africa = ['Nigeria', 'Egypt', 'South Africa']
    europe = ['France', 'Germany', 'Spain']
    asia = ['China', 'India', 'Japan']
    north_america = ['USA', 'Canada', 'Mexico']
    south_america = ['Brazil', 'Argentina', 'Colombia']
    oceania = ['Australia', 'New Zealand']
    
    if country in africa:
        return 'Africa'
    elif country in europe:
        return 'Europe'
    elif country in asia:
        return 'Asia'
    elif country in north_america:
        return 'North America'
    elif country in south_america:
        return 'South America'
    elif country in oceania:
        return 'Australia and Oceania'
    else:
        return 'Other'


df_clean['continent'] = df_clean['Country'].apply(result)
```

```python
import seaborn as sb
import pandas as pd
import matplotlib.pyplot as plt

data = {
    'category': ['A', 'B', 'A', 'C', 'B', 'A', 'B', 'C', 'd', 'e', 'f', 'g', 'd', 'e', 'f', 'g'],
    'number': [10, 20, 30, 40, 50, 60, 70, 80, 90, 20, 21, 22, 23, 24, 15, 26]
}
df_clean = pd.DataFrame(data)

base_color = sb.color_palette()[0]

category = df_clean['category'].value_counts().index

sb.countplot(data=df_clean, y='category', color=base_color, order=category)

plt.show()
```
